```python
import numpy as np                #Se importan los módulos que usaremos 
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
datos = np.genfromtxt('modulo-lunar.txt') # Se cargan los datos

#Se guardan los datos en arreglos de numpy
t = datos[:,0]    #Tiempo, en segundos. 
a = datos[:,1]    #Aceleración, en m/s^2
```

```python
#Definimos las constantes que se ocuparán en el código
m = 5000   #Kg, la masa del módulo
g = 9.8    #m/s^2, la acelaración de gravedad, que la supondremos constante
```

```python
plt.plot(t,a)               # Primer gráfico, datos observados
plt.grid(True)              # Ponemos una grilla en el gráfico, para ver mejor cómo cambian las variables.
plt.xlabel('$t\ \\rm [s]$') # Etiqueta del eje x, escrita en entorno matemático LaTeX
plt.ylabel('$a\ \\rm [m/s^2]$') # Etiqueta del eje y, escrita en entorno matemático LaTeX 
#plt.savefig('x_vs_t.png')
plt.show()
```

La segunda ley de Newton nos dice que
$$
\sum \vec{F} = m \vec{a}.
$$
Considerando el diagrama de fuerzas del enunciado, queda
$$
\vec{F_r} + \vec{F_g} = m_\text{m} \vec{a}, \quad (1)
$$
donde $\vec{F_g} = m_\text{m} \vec{g} $ es la fuerza de gravedad sobre el módulo, con $m_\text{m}$ la masa del módulo. 

Si consideramos el eje vertical de la figura como el eje $y$, con la coordenada positiva hacia arriba, tenemos entonces que $\vec{F_r} = F_r \, \hat{\jmath} $, y $\vec{F_g} =  m_\text{m} \vec{g} = m_\text{m} \, g  \, (- \hat{\jmath}) =  (- m_\text{m} \, g) \, \hat{\jmath}$. Reemplazando esto en (1), queda
$$
 F_r \, \hat{\jmath} + (- m_\text{m} \, g) \, \hat{\jmath} = (F_r - m_\text{m} \, g) \hat{\jmath} = m_\text{m} \vec{a}.
$$
Con esto, es claro que $\vec{a} = a \, \hat{\jmath}$, por lo que podemos escribir
$$
 F_r  - m_\text{m} \, g =   m_\text{m} a,
$$
de donde se puede obtener que
$$
F_r = m_\text{m}a + m_\text{m} g = m_\text{m} (a + g).
$$
Este resultado será usado en las siguentes celdas para graficar la fuera de roce.




```python
F_r = m * (a + g) 
plt.plot(t,F_r/1e6) #reescalado por un factor 10**6
plt.grid(True)
plt.xlabel('$t\ \\rm [s]$')
plt.ylabel('$F_r\ \\rm [10^6 N]$') #reescalado
#plt.ylim(0,0.2)
#plt.savefig('Dx_vs_t.png')
```

```python
F_r = m * (a + g) 
plt.plot(t,F_r/1e6) #reescalado por un factor 10**6
plt.grid(True)
plt.xlabel('$t\ \\rm [s]$')
plt.ylabel('$F_r\ \\rm [10^6 N]$') #reescalado
#plt.ylim(0,0.2)
#plt.savefig('Dx_vs_t.png')

plt.axvline(x = 5, linestyle='dashed') #Línea vertical mostrando el tiempo t = 5s
plt.axhline(y = 1, linestyle='dashed') #Línea horizontal mostrando la fuerza F_r = 10^6 N (considerando el reescalamiento)

plt.show()

```
