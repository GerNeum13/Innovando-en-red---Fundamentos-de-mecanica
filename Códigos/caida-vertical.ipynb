```python
import numpy as np                #Se importan los módulos que usaremos 
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
datos = np.genfromtxt('caida-vertical.txt') # Se cargan los datos
t = datos[:,0]                           # Se guardan los datos en arreglos de numpy
x = datos[:,1]
```

```python
plt.scatter(t,x)            # Primer gráfico, gráfico de los datos observados
plt.grid(True)              # Ponemos una grilla en el gráfico, paraver mejor como cambian las variables.
plt.xlabel('$t\ \\rm [s]$') # Etiqueta del eje x, escrita en entorno matemático latex
plt.ylabel('$y\ \\rm [m]$') # Etiqueta del eje y, escrita en entorno matemático latex (ponemos etiquetas similares en todso los gráficos)
plt.savefig('x_vs_t.png')
plt.show()
```

```python
Dx = np.diff(x)           # Cálculo de las diferencias de un punto al otro
Dt = np.diff(t)
```

```python
plt.scatter(t[:-1],Dx)                      # Gráfico de tiempo por desplazamiento en en tiempo. 
plt.grid(True)
plt.xlabel('$t\ \\rm [s]$')
plt.ylabel('$\Delta x\ \\rm [m]$')
#plt.ylim(0,0.2)
plt.savefig('Dx_vs_t.png')
plt.show()
```

```python
vm = Dx/Dt  # Cálculo de la velocidad por tramos
```

```python
plt.scatter(t[:-1],vm)      # Gráfico de velocidad por tiempo. 
plt.grid(True)
plt.xlabel('$t\ \\rm [s]$')
plt.ylabel('$\\bar{v}\ \\rm [m/s]$')
#plt.ylim(0,1)
plt.savefig('vm_vs_t.png')
plt.show()
```

```python
Dv = np.diff(vm)     # Cálculo de la diferencia de velocidad
am = Dv/Dt[-1]       # Cálculo de la aceleración por tramo
```

```python
plt.scatter(t[:-2],am)   # Gráfico de la aceleración pr tiempo
plt.grid(True)
plt.xlabel('$t\ \\rm [s]$')
plt.ylabel('$\\hat{a}\ \\rm [m/s^2]$')
plt.ylim(0,20)
plt.savefig('am_vs_t.png')
plt.show()
```

El resultado de este gráfico nos muestra que la aceleración es (más o menos) constante. Las desviaciones del valor constantes los atribuimos a errores experimentales. 

A partir del modelo idealizado,
$$
y(t) = y_0 + (g/2)t^2,
$$
podemos encontrar el valor de $y_0$ de modo que la curva pase por el primer punto de los datos, es decir $y(0.058 s) = -0.027 m$.

```python
#Se reemplaza el punto inicial de los datos en el modelo idealizado, y se despeja y_0 = y_0(t_0)
y_0 = -0.027 - (9.8/2)*(0.058)**2
```

```python
tt = np.linspace(0,0.7,1000)     #Se crea un arreglo de valores para la variable independiente.
y_teo = y_0 + (9.8/2)*tt**2      #Usamos el modelo idealizado para obtener los valores del eje y (distancia) para el arreglo de tiempos anterior
plt.plot(tt,y_teo)               #Graficamos el modelo idealizado
plt.scatter(t,x)                 #Graficamos los datos observados, para compararlos con el modelo idealizado.
plt.grid(True)
plt.xlabel('$t\ \\rm [s]$')
plt.ylabel('$y\ \\rm [m]$')
plt.savefig('x_vs_t.png')
plt.show()
```

Del gráfico anterior, se puede ver que el modelo idealizado ajusta bastante bien los datos observados.
